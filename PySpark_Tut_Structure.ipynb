# Step 3: Initialize Spark

```python
import os, findspark
os.environ["JAVA_HOME"] = r"D:\C\Program Files\Microsoft\jdk-11.0.22.7-hotspot"
findspark.init()

from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("SalesAnalysis").getOrCreate()
```

# Step 4: Create DataFrame

```python
sales_data = [
    (1, "Laptop", "Electronics", 1200, 2),
    (2, "Phone", "Electronics", 800, 1),
    (3, "Jeans", "Apparel", 50, 4),
    (4, "T-Shirt", "Apparel", 25, 10),
    (5, "Blender", "Home Appliances", 150, 1),
    (6, "Laptop", "Electronics", 1200, 1)
]
columns = ["Txn_ID", "Item", "Category", "Price", "Quantity"]
df = spark.createDataFrame(sales_data, schema=columns)
df.show()
```

# Step 5: Analysis

```python
from pyspark.sql.functions import sum, avg

# Total sales per category
df.groupBy("Category").agg(sum(df.Price * df.Quantity).alias("TotalSales")).show()

# Average price per category
df.groupBy("Category").agg(avg("Price").alias("AvgPrice")).show()
```
